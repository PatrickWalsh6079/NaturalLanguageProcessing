# Load Datasets

```python
import pandas as pd
import os

# set paths to each corpus
pos_path = 'C:/Users/Patrick/Syracuse_courses/IST_736/HW7/pos'
neg_path = 'C:/Users/Patrick/Syracuse_courses/IST_736/HW7/neg'

# create lists of each filename
pos_files = os.listdir(pos_path)
neg_files = os.listdir(neg_path)

print(pos_files[:10])
print(neg_files[:10])
```

```python
# create list of filenames with complete paths (positive files)
pos_filepaths = []
for file in pos_files:
    pos_filepaths.append(pos_path + '/' + file)
    
print(pos_filepaths[:10])
```

```python
# create list of filenames with complete paths (negative files)
neg_filepaths = []
for file in neg_files:
    neg_filepaths.append(neg_path + '/' + file)
    
print(neg_filepaths[:10])
```

# Data Preparation

### Tokenization & Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer

# create CV objects
CV1 = CountVectorizer(input='filename', stop_words='english')
CV2 = CountVectorizer(input='filename', stop_words='english')

# create document term matrices
pos_dtm = CV1.fit_transform(pos_filepaths)
neg_dtm = CV2.fit_transform(neg_filepaths)

print(pos_dtm[0])
print(neg_dtm[0])
```

```python
# extract column names
pos_colnames = CV1.get_feature_names_out()
print(pos_colnames)

neg_colnames = CV2.get_feature_names_out()
print(neg_colnames)
```

```python
# create dataframes with data from DTM and column names
pos_df = pd.DataFrame(pos_dtm.toarray(), columns=pos_colnames)
neg_df = pd.DataFrame(neg_dtm.toarray(), columns=neg_colnames)
```

```python
display(neg_df)
```

### Data Cleaning

```python
import re

def clean_cols(df):
    # drop columns with numbers
    columns_to_drop = [col for col in df.columns if any(char.isdigit() for char in col)]
    df = df.drop(columns = columns_to_drop)
    
    # drop columns with special characters
    pattern = '^[a-zA-Z]+$'  # regex pattern to match only letters from a-z or A-Z
    columns_to_drop = [col for col in df.columns if not re.match(pattern, col)]
    df = df.drop(columns = columns_to_drop)
    
    # drop columns that have a length of less than 3 or more than 15
    columns_to_drop = [col for col in df.columns if len(col) < 3 or len(col) > 15]
    df = df.drop(columns = columns_to_drop)
    
    # drop columns that have three or more consecutive occurrences of the same letter
    pattern_consecutive = r'([a-zA-Z])\1\1'
    columns_to_drop = [col for col in df.columns if re.search(pattern_consecutive, col)]
    df = df.drop(columns=columns_to_drop)
    
    return df

cleaned_pos_df = clean_cols(pos_df)
display(cleaned_pos_df)
```

```python
cleaned_neg_df = clean_cols(neg_df)
display(cleaned_neg_df)
```

```python
# see complete list of column names
for col in cleaned_pos_df.columns:
    print(col)
```

```python
# add LABEL column
cleaned_neg_df.insert(0, 'LABEL', 'neg')
display(cleaned_neg_df)
```

```python
# combine dataframes into one dataframe
combined_df = pd.concat([cleaned_pos_df, cleaned_neg_df], ignore_index=True)
display(combined_df)
```

```python
# replace null values with 0
combined_df.fillna(0, inplace=True)
display(combined_df)
```

# Train Models
### Train/Test Split

```python
from sklearn.model_selection import train_test_split

# create training and testing sets from dataframe
TrainSet, TestSet = train_test_split(combined_df, test_size=0.3)

# split testing labels from testing data
TestLabels = TestSet['LABEL']
TestSet = TestSet.drop(["LABEL"], axis=1)  # remove the entire column

# split training labels from training data
TrainLabels = TrainSet['LABEL']
TrainSet = TrainSet.drop(["LABEL"], axis=1)  # remove the entire column
```

### Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
import numpy as np
from tqdm import tqdm  # Import tqdm for progress bar
import time  # Import time module

# Initialize MultinomialNB model
MyModel = MultinomialNB()

# Start the timer
start_time = time.time()

# Fit the model with progress bar
with tqdm(total=len(TrainSet)) as pbar:
    trainedMNB = MyModel.fit(TrainSet, TrainLabels)
    pbar.update(len(TrainSet))

# Predict with the trained model
with tqdm(total=len(TestSet)) as pbar:
    predMNB = MyModel.predict(TestSet)
    pbar.update(len(TestSet))

# Print the predictions
print(predMNB)
print(np.round(MyModel.predict_proba(TestSet), 2))

# Calculate the time taken
end_time = time.time()
elapsed_time = end_time - start_time

# Convert elapsed time to minutes and seconds
minutes = int(elapsed_time // 60)
seconds = int(elapsed_time % 60)

# Print the time taken
print(f"Time taken: {minutes} minutes and {seconds} seconds")
```

### Support Vector Machines (SVM)

```python
from sklearn.preprocessing import MinMaxScaler

# scale training and testing sets for use by SVMs
scaling = MinMaxScaler(feature_range=(-1,1)).fit(TrainSet)
TrainSetScaled = scaling.transform(TrainSet)
TestSetScaled = scaling.transform(TestSet)
```

```python
from sklearn.svm import LinearSVC

# Initialize LinearSVC model
SVM_Model1 = LinearSVC(C=10, 
                       max_iter=10000,
                       dual=True,
                       verbose=True)

# Start the timer
start_time = time.time()

# Fit the model with progress bar
with tqdm(total=len(TrainSetScaled)) as pbar:
    trainedSVM1 = SVM_Model1.fit(TrainSetScaled, TrainLabels)
    pbar.update(len(TrainSetScaled))

# Predict with the trained model
with tqdm(total=len(TestSetScaled)) as pbar:
    predSVM1 = SVM_Model1.predict(TestSetScaled)
    pbar.update(len(TestSetScaled))

# Print the predictions
print(predSVM1)

# Calculate the time taken
end_time = time.time()
elapsed_time = end_time - start_time

# Convert elapsed time to minutes and seconds
minutes = int(elapsed_time // 60)
seconds = int(elapsed_time % 60)

# Print the time taken
print(f"Time taken: {minutes} minutes and {seconds} seconds")
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Initialize the pipeline with StandardScaler and SGDClassifier
SVM_Model4 = make_pipeline(StandardScaler(),
                           SGDClassifier(max_iter=10000, tol=1e-3))

# Start the timer
start_time = time.time()

# Fit the model with progress bar
with tqdm(total=len(TrainSet)) as pbar:
    trainedSVM4 = SVM_Model4.fit(TrainSet, TrainLabels)
    pbar.update(len(TrainSet))

# Predict with the trained model
predSVM4 = SVM_Model4.predict(TestSet)

# Calculate the time taken
end_time = time.time()
elapsed_time = end_time - start_time

# Convert elapsed time to minutes and seconds
minutes = int(elapsed_time // 60)
seconds = int(elapsed_time % 60)

# Print the predictions
print(predSVM4)

# Print the time taken
print(f"Time taken: {minutes} minutes and {seconds} seconds")
```

### Feature Importance

```python
print('MODEL 1')
# Get the feature log probabilities
feature_log_probs = MyModel.feature_log_prob_

# Get feature names
feature_names = TrainSet.columns

# Find the indices of the top 10 most important words for negative and positive classes
neg_class_indices = np.argsort(feature_log_probs[0, :])[-10:][::-1]
pos_class_indices = np.argsort(feature_log_probs[1, :])[-10:][::-1]

# Print the top 10 most important words for negative class
print("Top 10 most important words for negative class:")
for i in neg_class_indices:
    print(feature_names[i])

# Print the top 10 most important words for positive class
print("\nTop 10 most important words for positive class:")
for i in pos_class_indices:
    print(feature_names[i])
```
