# Tokenize & Vectorize

```python
from sklearn.feature_extraction.text import CountVectorizer

# create CV objects
CV = CountVectorizer(input='filename', stop_words='english', encoding='latin-1')

# create document term matrix
dtm = CV.fit_transform(master_list)

print(dtm[0])
```

```python
# extract column names
colnames = CV.get_feature_names_out()
print(len(colnames))
print(colnames)
```

```python
# create dataframes with data from DTM and column names
df = pd.DataFrame(dtm.toarray(), columns=colnames)
display(df.head())
```

# Data Cleaning

```python
import re

def clean_cols(df):
    # drop columns with numbers
    columns_to_drop = [col for col in df.columns if any(char.isdigit() for char in col)]
    df = df.drop(columns = columns_to_drop)
    
    # drop columns with special characters
    pattern = '^[a-zA-Z]+$'  # regex pattern to match only letters from a-z or A-Z
    columns_to_drop = [col for col in df.columns if not re.match(pattern, col)]
    df = df.drop(columns = columns_to_drop)
    
    # drop columns that have a length of less than 3 or more than 15
    columns_to_drop = [col for col in df.columns if len(col) < 3 or len(col) > 15]
    df = df.drop(columns = columns_to_drop)
    
    # drop columns that have three or more consecutive occurrences of the same letter
    pattern_consecutive = r'([a-zA-Z])\1\1'
    columns_to_drop = [col for col in df.columns if re.search(pattern_consecutive, col)]
    df = df.drop(columns=columns_to_drop)
    
    return df

cleaned_df = clean_cols(df)
display(cleaned_df)
```

```python
# see complete list of column names
for col in cleaned_df.columns:
    print(col)
```

### Update CountVectorizer and Document Term Matrix after cleaning

```python
# Update CountVectorizer with updated column names
CV = CountVectorizer(input='content', stop_words='english', encoding='latin-1', vocabulary=cleaned_df.columns)

# Convert cleaned DataFrame to a list of strings
cleaned_texts = cleaned_df.apply(lambda row: ' '.join(row.values.astype(str)), axis=1).tolist()

# Transform cleaned texts using updated CountVectorizer
dtm = CV.fit_transform(cleaned_texts)
```

```python
from sklearn.decomposition import LatentDirichletAllocation
from tqdm import tqdm
import time
import pyLDAvis.lda_model
import pyLDAvis
import matplotlib.pyplot as plt
import numpy as np
    

def run_LDA(df, num_topics, colnames):
    lda_model = LatentDirichletAllocation(n_components=num_topics, max_iter=100, learning_method='online')

    # Start the timer
    start_time = time.time()
    
    # Fit the model with progress bar
    with tqdm(total=len(df)) as pbar:
        lda_fitted = lda_model.fit_transform(df)
        pbar.update(len(df))

    # Print the time taken
    end_time = time.time()
    elapsed_time = end_time - start_time
    minutes = int(elapsed_time // 60)
    seconds = int(elapsed_time % 60)
    print(f"Time taken: {minutes} minutes and {seconds} seconds")

    # Get the top words for each topic
    num_words = 10
    for idx, topic in enumerate(lda_model.components_):
        print("Topic %d:" % (idx))
        topic_words = [(word, topic[word_id]) for word_id, word in enumerate(colnames[:len(topic)])]
        topic_words_sorted = sorted(topic_words, key=lambda x: x[1], reverse=True)[:num_words]
        print(topic_words_sorted)

    pyLDAvis.enable_notebook()
    panel = pyLDAvis.lda_model.prepare(lda_model, dtm, CV, mds='tsne')
    pyLDAvis.save_html(panel, f"News with {NUM_TOPICS} topics.html")
    

    # show word counts per topic
    word_topic = np.array(lda_model.components_)
    word_topic = word_topic.transpose()

    num_top_words = 15
    vocab_array = np.asarray(colnames)

    fontsize_base = 10

    # Calculate the number of rows and columns based on NUM_TOPICS
    num_topics = word_topic.shape[1]
    num_cols = min(5, num_topics)  # Maximum of 5 columns
    num_rows = (num_topics + num_cols - 1) // num_cols  # Ceiling division to ensure enough rows

    fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 3))  # Adjust figsize as needed

    for t in range(num_topics):
        row = t // num_cols  # Calculate the row index
        col = t % num_cols   # Calculate the column index

        if num_rows > 1:
            ax = axs[row, col]
        else:
            ax = axs[col]  # Handle single-row case

        ax.set_ylim(0, num_top_words + 0.5)
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_title('Topic #{}'.format(t))

        top_words_idx = np.argsort(word_topic[:, t])[::-1][:num_top_words]
        top_words = vocab_array[top_words_idx]
        top_words_shares = word_topic[top_words_idx, t]

        for i, (word, share) in enumerate(zip(top_words, top_words_shares)):
            ax.text(0.1, num_top_words - i - 0.5, word, fontsize=fontsize_base)

    plt.tight_layout()
    plt.show()
```
